```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def phi(x, a, b, c):
    return (a / 2) / (1 + np.cosh((x - b) / c))
    
def phiphi(x, y, a, b, c):
    return phi(x, a, b, c) * phi(-y, a, b, c)
```

```python
# a = np.asarray([0.113, 2.224e-3, 1.812e-2])
a = np.asarray([1, 0, 0])
# b = np.asarray([56.468, 30.197, 86.148])
b = np.asarray([5, 0, 0])
# c = np.asarray([6.035, 284.54, 38.842])
c = np.asarray([1, 1, 1])
```

```python
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
```

```python
Z = phiphi(X, Y, a[0], b[0], c[0]) + phiphi(X, Y, a[1], b[1], c[1]) + phiphi(X, Y, a[2], b[2], c[2])
Z = np.where(X < Y, Z, np.nan)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.tight_layout()
plt.show()
```
