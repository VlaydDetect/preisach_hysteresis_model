```python
def matrix_exponent(A: np.ndarray, t: float, b: np.ndarray) -> np.ndarray:
    e = np.linalg.inv(np.identity(2) - expm(2.0 * np.pi * A))
    # print(e)
    e = e @ expm(A * t)
    # print(e)
    e = e @ b
    # print(e)
    return e
```

```python
def create_hurwitz_matrix(n):
    """
    Создаёт случайную гурвицеву матрицу n x n.
    """
    # # Создаём случайную матрицу
    # random_matrix = np.random.rand(n, n)
    # # Создаём кососимметричную матрицу (собственные значения на мнимой оси)
    # skew_symmetric = random_matrix - random_matrix.T
    # # Сдвигаем собственные значения влево, чтобы сделать их действительные части отрицательными
    # A = skew_symmetric - np.eye(n) * 0.5
    # print("Сгенерированная гурвицева матрица A:\n", A)
    # print("Собственные значения A:", np.linalg.eigvals(A))
    A = np.array([
        [0.0, 1.0],
        [-0.1, -0.7]
    ])
    print("матрица A:\n", A)
    print("Собственные значения A:", np.linalg.eigvals(A))
    return A
```

```python
def plot_corporeal_cone(n_dim, t_max=None, num_points=None):
    """
    Строит и отображает телесный конус для системы размерности n_dim (2 или 3).
    """
    if n_dim not in [2, 3]:
        raise ValueError("Размерность должна быть 2 или 3.")

    # 1. Определение параметров системы
    A = create_hurwitz_matrix(n_dim)
    # b = np.random.rand(n_dim) * 2 - 1  # Случайный вектор b
    b = np.array([0.0, 1.5])
    print("\nВектор b:\n", b)

    # 2. Генерация образующей кривой
    if t_max is None:
        t_max = 15
    if num_points is None:
        num_points = 200
    t_space = np.linspace(0, t_max, num_points, dtype=np.float64)

    # Вычисляем траекторию x(t) = exp(A*t) * b
    trajectory_points = np.array([matrix_exponent(A, t, b) for t in t_space])
    # trajectory_points = np.array([matrix_exponent(A, t, b, c) for t in t_space])

    # 3. Формирование точек для выпуклой оболочки.
    # Добавляем начало координат (вершину конуса)
    origin = np.zeros((1, n_dim))
    points_for_hull = np.vstack([origin, trajectory_points])

    # 4. Вычисление выпуклой оболочки
    hull = ConvexHull(points_for_hull)

    # 5. Визуализация
    if n_dim == 2:
        fig, ax = plt.subplots(figsize=(8, 8))
        ax.set_title("Телесный конус (2D)", fontsize=16)

        # Рисуем траекторию
        ax.plot(trajectory_points[:, 0], trajectory_points[:, 1], 'b-', label='Траектория $e^{At}b$')

        # Рисуем вершину конуса
        ax.plot(0, 0, 'ko', label='Начало координат (вершина)')

        samples_count = 100
        tangents = np.array([])
        for i in range(samples_count):
            ti = i / (samples_count - 1) * t_max
            print(ti)
            E = matrix_exponent(A, ti, b)
            print(E)
            x0 = np.random.random(2)
            x0 /= np.linalg.norm(x0)
            print(x0)
            g = E @ x0
            print(g)
            tg = A @ g
            print(tg)
            norm = np.linalg.norm(tg)
            if norm > 1e-9 and np.isfinite(norm):
                tg /= norm
                ax.plot(tg, label='rays')
                # tangents = np.append(tangents, tg)

        # Рисуем выпуклую оболочку (конус)
        # for simplex in hull.simplices:
        #     ax.plot(points_for_hull[simplex, 0], points_for_hull[simplex, 1], 'r-', alpha=0.8)
        # 
        # # Заливаем конус для наглядности
        # hull_polygon = Polygon(points_for_hull[hull.vertices], color='red', alpha=0.15)
        # ax.add_patch(hull_polygon)

        ax.set_xlabel("x1")
        ax.set_ylabel("x2")
        ax.grid(True)
        ax.legend()
        ax.axhline(y=0, color='k', lw=0.8)
        ax.axvline(x=0, color='k', lw=0.8)
        ax.axis('equal')

    elif n_dim == 3:
        fig = plt.figure(figsize=(10, 10))
        ax = fig.add_subplot(111, projection='3d')
        ax.set_title("Телесный конус (3D)", fontsize=16)

        # Рисуем траекторию
        ax.plot(trajectory_points[:, 0], trajectory_points[:, 1], trajectory_points[:, 2], 'b-',
                label='Траектория $e^{At}b$')

        # Рисуем вершину
        ax.scatter(0, 0, 0, c='k', marker='o', s=50, label='Начало координат (вершина)')

        # Рисуем поверхность конуса
        # hull.simplices содержит индексы треугольников, образующих поверхность
        ax.plot_trisurf(
            points_for_hull[:, 0], points_for_hull[:, 1], points_for_hull[:, 2],
            triangles=hull.simplices,
            color='red', alpha=0.2, shade=True
        )

        ax.set_xlabel("x1")
        ax.set_ylabel("x2")
        ax.set_zlabel("x3")
        ax.legend()

    plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from scipy.linalg import expm
from scipy.spatial import ConvexHull
```

```python
def plot_plotly_3d_cone():
    """
    Строит и отображает вращаемый 3D телесный конус с помощью Plotly.
    """
    # 1. Параметры системы
    n_dim = 3
    A = create_hurwitz_matrix(n_dim)
    b = np.random.rand(n_dim) * 2 - 1

    # 2. Генерация траектории
    t_space = np.linspace(0, 15, 200)
    trajectory_points = np.array([expm(A * t) @ b for t in t_space])

    # 3. Точки для выпуклой оболочки
    origin = np.zeros((1, n_dim))
    points_for_hull = np.vstack([origin, trajectory_points])

    # 4. Вычисление выпуклой оболочки
    hull = ConvexHull(points_for_hull)

    # 5. Визуализация с Plotly

    # Данные для траектории
    trace_trajectory = go.Scatter3d(
        x=trajectory_points[:, 0],
        y=trajectory_points[:, 1],
        z=trajectory_points[:, 2],
        mode='lines',
        line=dict(color='blue', width=4),
        name='Траектория e^(At)b'
    )

    # Данные для поверхности конуса
    # Plotly использует Mesh3d для отрисовки полигональных сеток
    trace_cone = go.Mesh3d(
        x=points_for_hull[:, 0],
        y=points_for_hull[:, 1],
        z=points_for_hull[:, 2],
        # i, j, k задают треугольники поверхности из hull.simplices
        i=hull.simplices[:, 0],
        j=hull.simplices[:, 1],
        k=hull.simplices[:, 2],
        color='red',
        opacity=0.20,
        name='Телесный конус K'
    )

    # Создание фигуры и компоновка
    fig = go.Figure(data=[trace_trajectory, trace_cone])

    fig.update_layout(
        title_text='Интерактивный телесный конус (Plotly)',
        scene=dict(
            xaxis_title='x1',
            yaxis_title='x2',
            zaxis_title='x3'
        )
    )

    fig.show()
```

```python
plot_plotly_3d_cone()
```

```python
# Solid cone visualization for x' = A x with input b u(t), using the curve {exp(At) b, t >= 0}
# We avoid SciPy to stay self-contained: compute x(t) by integrating x' = A x with RK4.
# For n>3, we project to 3D with PCA for visualization.

import numpy as np
import matplotlib.pyplot as plt


def rk4_step(f, t, y, h):
    k1 = f(t, y)
    k2 = f(t + 0.5 * h, y + 0.5 * h * k1)
    k3 = f(t + 0.5 * h, y + 0.5 * h * k2)
    k4 = f(t + h, y + h * k3)
    return y + (h / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)


def generate_curve_expAt_b(A, b, dt=0.01, t_max=20.0, tol=1e-6, sample_every=10):
    """
    Integrate x' = A x, x(0)=b, collect points x(t)=exp(At) b.
    Stop when ||x(t)|| < tol or t exceeds t_max.
    """
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float).reshape(-1)
    n = A.shape[0]
    assert A.shape == (n, n)
    assert b.shape == (n,)

    def f(_, x):
        return A @ x

    t = 0.0
    x = b.copy()
    pts = [x.copy()]
    ts = [t]
    steps = 0
    while t < t_max and np.linalg.norm(x) > tol:
        x = rk4_step(f, t, x, dt)
        t += dt
        steps += 1
        if steps % sample_every == 0:
            pts.append(x.copy())
            ts.append(t)
    # ensure last point is included
    if len(ts) == 0 or ts[-1] < t:
        pts.append(x.copy())
        ts.append(t)
    return np.array(ts), np.vstack(pts)


def pca_project_to_3d(P):
    # Center then take top-3 singular vectors
    P = np.asarray(P, dtype=float)
    mean = P.mean(axis=0, keepdims=True)
    X = P - mean
    U, S, Vt = np.linalg.svd(X, full_matrices=False)
    V3 = Vt[:3].T  # n x 3
    P3 = X @ V3
    return P3, V3, mean.ravel()


def plot_cone_from_curve(P, title="Solid cone (approx.) from {exp(At) b}"):
    """
    P: (m, n) points along the curve exp(At) b, ordered by increasing t (towards the origin).
    If n == 2: plot 2D fan.
    If n >= 3: project to 3D by PCA then plot triangular fan from origin to consecutive points.
    """
    m, n = P.shape
    if n == 1:
        # Degenerate 1D: just a line
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.plot(np.arange(m), P[:, 0])
        ax.set_title(title + " (1D)")
        ax.set_xlabel("sample index")
        ax.set_ylabel("x")
        plt.show()
        return

    if n == 2:
        # 2D: draw origin and polygonal chain plus filled fan
        O = np.zeros(2)
        fig, ax = plt.subplots(figsize=(6, 6))
        # fan fill: triangles (O, P[i], P[i+1]) approximated by a filled polygon per segment
        for i in range(m - 1):
            tri = np.vstack([O, P[i], P[i + 1]])
            ax.fill(tri[:, 0], tri[:, 1], alpha=0.2, linewidth=0)
        ax.plot(P[:, 0], P[:, 1], linewidth=1.0)
        ax.scatter([0], [0], marker='o')
        ax.set_aspect('equal', 'box')
        ax.set_title(title + " (2D)")
        ax.set_xlabel("x1")
        ax.set_ylabel("x2")
        ax.grid(True, linestyle=':')
        plt.show()
        return

    # n >= 3: project to 3D
    P3, _, _ = pca_project_to_3d(P)
    from mpl_toolkits.mplot3d.art3d import Poly3DCollection
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(111, projection='3d')
    O = np.zeros(3)
    # Build triangular fan surface from origin
    faces = []
    for i in range(m - 1):
        faces.append([O, P3[i], P3[i + 1]])
    poly = Poly3DCollection(faces, alpha=0.25)
    ax.add_collection3d(poly)
    # Plot the curve itself
    ax.plot(P3[:, 0], P3[:, 1], P3[:, 2], linewidth=1.0)
    ax.scatter([0], [0], [0], marker='o')
    # Set equal-ish aspect
    max_range = (P3.max(axis=0) - P3.min(axis=0)).max()
    mid = P3.mean(axis=0)
    for dim, mval in enumerate(mid):
        getattr(ax, f"set_{'xyz'[dim]}lim")(mval - 0.6 * max_range, mval + 0.6 * max_range)
    ax.set_title(title + " (PCA to 3D)")
    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")
    ax.set_zlabel("PC3")
    plt.show()


# --- Demo with a 3x3 Hurwitz matrix A and a vector b ---
np.random.seed(0)

# Construct a random Hurwitz matrix via symmetric part with negative definite property
# Q = np.random.randn(3,3)
# A_demo = -0.5*np.eye(3) + 0.2*(Q - Q.T)  # stable (Hurwitz) with some rotation
# b_demo = np.array([1.0, 0.2, -0.3])

A_demo = np.array([
    [0.0, 1.0],
    [-1.0, -0.1]
])
b_demo = np.array([0.0, 1.5])

ts, P = generate_curve_expAt_b(A_demo, b_demo, dt=0.01, t_max=25.0, tol=1e-6, sample_every=10)

plot_cone_from_curve(P, title="Solid cone for x' = A x, x(0)=b")
```

```python
def plot_curve_and_eigenvectors(A, b, t_max=2.0, n_points=200):
    """
    Строит кривую exp(A*t)*b и лучи вдоль собственных векторов матрицы A.

    Args:
        A (np.ndarray): Квадратная матрица n x n.
        b (np.ndarray): Начальный n-мерный вектор.
        t_max (float): Максимальное значение времени t.
        n_points (int): Количество точек для построения кривой.
    """
    n = A.shape[0]
    if n not in [2, 3]:
        print("Визуализация поддерживается только для 2D и 3D случаев.")
        return

    # 1. Вычисляем собственные векторы и значения
    eigenvalues, eigenvectors = np.linalg.eig(A)
    print(f"Матрица A ({n}x{n}):\n{A}")
    print(f"\nСобственные значения: {eigenvalues}")
    print(f"\nСобственные векторы (по столбцам):\n{eigenvectors}")

    # 2. Генерируем точки кривой
    t_space = np.linspace(0, t_max, n_points)
    curve_points = np.array([expm(A * t) @ b for t in t_space])

    # 3. Настраиваем график
    fig = plt.figure(figsize=(10, 8))
    if n == 2:
        ax = fig.add_subplot(111)
        ax.set_title("2D случай: Кривая и собственные векторы", fontsize=16)
        ax.set_xlabel("X")
        ax.set_ylabel("Y")
    else:  # n == 3
        ax = fig.add_subplot(111, projection='3d')
        ax.set_title("3D случай: Кривая и собственные векторы", fontsize=16)
        ax.set_xlabel("X")
        ax.set_ylabel("Y")
        ax.set_zlabel("Z")

    # 4. Строим кривую
    if n == 2:
        ax.plot(curve_points[:, 0], curve_points[:, 1], 'b-', linewidth=2.5, label='Кривая exp(At)b')
        # Начальная точка
        ax.plot(b[0], b[1], 'ro', markersize=8, label='Начальная точка b')
    else:  # n == 3
        ax.plot(curve_points[:, 0], curve_points[:, 1], curve_points[:, 2], 'b-', linewidth=2.5,
                label='Кривая exp(At)b')
        # Начальная точка
        ax.plot([b[0]], [b[1]], [b[2]], 'ro', markersize=8, label='Начальная точка b')

    # 5. Строим лучи (собственные векторы)
    # Определяем масштаб для лучей, чтобы они были хорошо видны
    max_val = np.max(np.abs(curve_points)) * 1.2

    for i in range(n):
        vec = eigenvectors[:, i].real  # Берем реальную часть на случай малых вычислительных ошибок
        # Строим луч в положительном и отрицательном направлениях от начала координат
        ray_points = np.array([-vec * max_val, vec * max_val])
        if n == 2:
            ax.plot(ray_points[:, 0], ray_points[:, 1], 'k--', linewidth=1.5,
                    label=f'Собств. вектор для λ={eigenvalues[i].real:.2f}')
        else:  # n == 3
            ax.plot(ray_points[:, 0], ray_points[:, 1], ray_points[:, 2], 'k--', linewidth=1.5,
                    label=f'Собств. вектор для λ={eigenvalues[i].real:.2f}')

    ax.grid(True)
    ax.legend()
    # Для 2D делаем оси равными
    if n == 2:
        ax.set_aspect('equal', adjustable='box')
    plt.show()
```
