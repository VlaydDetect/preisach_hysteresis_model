```python
import numpy as np
from scipy.linalg import expm
from scipy.spatial import ConvexHull
from scipy.optimize import nnls
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
def sample_curve(A, b, T=10.0, num=1001):
    t = np.linspace(0, T, num)
    X = np.stack([expm(A * tt) @ b for tt in t], axis=0)  # shape (num, n)
    return t, X


def normalize_rows(X, eps=1e-12):
    norms = np.linalg.norm(X, axis=1, keepdims=True)
    norms = np.maximum(norms, eps)
    return X / norms, norms.ravel()
```

```python
def hull_vertices_on_sphere(unit_dirs):
    # Compute convex hull of points on sphere in ambient R^n and return vertex indices
    if unit_dirs.shape[0] <= unit_dirs.shape[1] + 1:
        # too few points: take all
        return np.arange(unit_dirs.shape[0])
    try:
        hull = ConvexHull(unit_dirs)
        verts = np.unique(hull.vertices)
        return verts
    except Exception:
        # fallback: all indices
        return np.arange(unit_dirs.shape[0])
```

```python
def select_n_rays_from_vertices(unit_dirs, vert_idx, n_select):
    V = unit_dirs[vert_idx]  # candidates
    m = V.shape[0]
    if m <= n_select:
        return V.copy(), vert_idx.copy()
    # Greedy farthest sampling on sphere (maximize minimal angular separation)
    selected = []
    selected_idx = []
    # start from the vertex most aligned with mean direction
    mean_dir = np.mean(V, axis=0)
    i0 = np.argmax(V @ mean_dir)
    selected.append(V[i0])
    selected_idx.append(vert_idx[i0])
    remaining = list(range(m))
    remaining.remove(i0)
    for _ in range(1, n_select):
        best = None
        best_val = -1.0
        for j in remaining:
            v = V[j]
            # measure distance as minimal (1 - abs(dot)) to already selected
            dots = np.abs(np.dot(np.stack(selected), v))
            min_dist = np.min(1.0 - dots)  # larger -> more distant
            if min_dist > best_val:
                best_val = min_dist
                best = j
        selected.append(V[best])
        selected_idx.append(vert_idx[best])
        remaining.remove(best)
    return np.vstack(selected), np.array(selected_idx)
```

```python
def check_inclusion_by_nnls(unit_dirs, rays):
    # rays: shape (k, n) -> transpose to columns (n, k)
    R = rays.T  # shape (n, k)
    max_resid = 0.0
    resid_list = []
    for u in unit_dirs:
        c, rnorm = nnls(R, u)
        # compute residual ||R c - u||
        resid = np.linalg.norm(R @ c - u)
        resid_list.append(resid)
        if resid > max_resid:
            max_resid = resid
    return max_resid, np.array(resid_list)
```

```python
# --- Visualization helpers ---
def plot_2d_curve_and_cone(X, rays, title="2D: curve and cone"):
    plt.figure(figsize=(6, 6))
    plt.plot(X[:, 0], X[:, 1], '-', lw=2, label='curve x(t)')
    max_norm = np.max(np.linalg.norm(X, axis=1))
    # plot rays
    scaled = rays * max_norm * 1.05
    for i, r in enumerate(scaled):
        plt.plot([0, r[0]], [0, r[1]], '--', lw=2, label=f'ray {i + 1}')
    # Shade cone region (for 2 rays only)
    if rays.shape[0] == 2:
        # polygon: origin, r1, r2
        poly_x = [0, scaled[0, 0], scaled[1, 0]]
        poly_y = [0, scaled[0, 1], scaled[1, 1]]
        plt.fill(poly_x, poly_y, alpha=0.2)
    plt.gca().set_aspect('equal', 'box')
    plt.grid(True)
    plt.legend()
    plt.title(title)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.show()
```

```python
def plot_3d_curve_and_cone(X, rays, title="3D: curve and cone"):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(X[:, 0], X[:, 1], X[:, 2], '-', lw=2, label='curve x(t)')
    max_norm = np.max(np.linalg.norm(X, axis=1))
    scaled = rays * max_norm * 1.05
    # plot rays
    for i, r in enumerate(scaled):
        ax.plot([0, r[0]], [0, r[1]], [0, r[2]], '--', lw=2, label=f'ray {i + 1}')
    # if we have 3 rays, draw triangular cone faces (origin, r_i, r_j)
    if rays.shape[0] == 3:
        faces = [
            [[0, 0, 0], scaled[0], scaled[1]],
            [[0, 0, 0], scaled[1], scaled[2]],
            [[0, 0, 0], scaled[2], scaled[0]],
        ]
        poly = Poly3DCollection(faces, alpha=0.15)
        ax.add_collection3d(poly)
    # adjust axes limits
    all_pts = np.vstack([X, scaled])
    mins = all_pts.min(axis=0)
    maxs = all_pts.max(axis=0)
    for i in range(3):
        ax.set_xlim(mins[0] * 1.1, maxs[0] * 1.1)
        ax.set_ylim(mins[1] * 1.1, maxs[1] * 1.1)
        ax.set_zlim(mins[2] * 1.1, maxs[2] * 1.1)
    ax.set_box_aspect([1, 1, 1])
    ax.legend()
    ax.set_title(title)
    plt.show()
```

```python
# --- Example runs for 2D and 3D ---

def run_example_2d():
    n = 2
    # choose A with real eigenvalues (distinct) and non-trivial off-diagonal to make curve interesting
    A = np.array(
        [[-1.0, 0.8],
         [0.0, -2.0]]
    )
    b = np.array([1.0, 0.6])
    
    A2 = np.array([
        [0.0, 1.0],
        [-0.1, -0.7]
    ])
    b2 = np.array([1, -1])
    
    T = 100.0
    t, X = sample_curve(A2, b2, T=T, num=20001)
    unit_dirs, norms = normalize_rows(X)
    verts_idx = hull_vertices_on_sphere(unit_dirs)
    rays, sel_idx = select_n_rays_from_vertices(unit_dirs, verts_idx, n_select=2)
    max_resid, resid_list = check_inclusion_by_nnls(unit_dirs, rays)
    print("2D example:")
    print("Selected ray indices (in samples):", sel_idx)
    print("Max NNLS residual (should be small if cone contains directions):", max_resid)
    plot_2d_curve_and_cone(X, rays, title="2D: exp(A t) b and cone (2 rays)")
```

```python
run_example_2d()
```

```python
def run_example_3d():
    n = 3
    # choose A with real eigenvalues (diagonal with different decays), add mild coupling via similarity transform
    D = np.diag([-0.6, -1.2, -2.0])
    # apply a similarity transform to make A non-diagonal but keep real eigenvalues
    Q = np.array([[1.0, 0.3, 0.2],
                  [0.2, 1.0, 0.4],
                  [0.1, 0.2, 1.0]])
    A = np.linalg.inv(Q) @ D @ Q
    b = np.array([1.0, 0.5, 0.2])
    T = 10.0
    t, X = sample_curve(A, b, T=T, num=2001)
    unit_dirs, norms = normalize_rows(X)
    verts_idx = hull_vertices_on_sphere(unit_dirs)
    rays, sel_idx = select_n_rays_from_vertices(unit_dirs, verts_idx, n_select=3)
    max_resid, resid_list = check_inclusion_by_nnls(unit_dirs, rays)
    print("3D example:")
    print("Selected ray indices (in samples):", sel_idx)
    print("Max NNLS residual (should be small if cone contains directions):", max_resid)
    plot_3d_curve_and_cone(X, rays, title="3D: exp(A t) b and cone (3 rays)")
```

```python
run_example_3d()
```
